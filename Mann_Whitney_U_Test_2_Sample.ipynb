```python
# https://statistics.laerd.com/spss-tutorials/mann-whitney-u-test-using-spss-statistics.php
# 1. H0: The MEAN mpg is equal between the two groups
# 2. H1: The MEAN mpg is not equal between the two groups
# Confidence Level 95%   
group1 = [20, 23, 21, 25, 18, 17, 18, 24, 20, 24, 23, 19]
group2 = [24, 25, 21, 22, 23, 18, 17, 28, 24, 27, 21, 23]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
sns.distplot(group1)
```

```python
sns.distplot(group2)
```

```python
# We can see from above that data does not come from a normal distributed population
# Shapes.
# If they similar shape, the proceed with mann whitney the compare medians.(mean ranks)
# If they do not have similar shape, the proceed with mann whitney compare the means
from scipy.stats import mannwhitneyu
stats, pvalue = mannwhitneyu(group1, group2, alternative = 'two-sided')
print('P value ', pvalue)
if pvalue < 0.05:
   print('Reject the null, conclude The MEAN mpg is not equal between the two groups')
else:
   print('Fail to Reject Null, The MEAN mpg is equal between the two groups')
```
