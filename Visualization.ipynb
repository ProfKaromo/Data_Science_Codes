```python
import pandas
data = pandas.read_csv("https://modcom.co.ke/data/datasets/schoolcleaned.csv", 
                       parse_dates=['bday','enrolldate', 'expgradate'])
data.head(4)
```

```python
#Matplotlib, Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.available
```

```python
plt.style.use('ggplot')
#count plot, 
```

```python
sns.countplot(data['Smoking'], palette='gnuplot2')
plt.title('Smoking Counts')
plt.xlabel('Smoking Students')
plt.ylabel('Smoking Students Count')
```

```python
#Bar plot
import numpy
ax = sns.barplot(x='Gender', y='Math', data=data, estimator= numpy.mean, palette='Oranges', hue='Athlete')
for p in ax.patches:
  txt = str(p.get_height().round(1)) + ' '
  txt_x = p.get_x()
  txt_y = p.get_height()
  ax.text(txt_x,txt_y,txt)
```

```python
#Dist plot
#sns.distplot(data['Math'])
sns.distplot(data['Math'], kde=False)
# for p in ax.patches:
#  txt = str(p.get_height().round(1)) + ' '
#  txt_x = p.get_x()
#  txt_y = p.get_height()
#  ax.text(txt_x,txt_y,txt)
plt.title('Distribution Graph')
plt.xlabel('Maths')
plt.ylabel('Frequency')
```

```python
#Swam plot
color = ['#B388FF','#64DD17','#E65100']
#sns.swarmplot(x='Gender', y='Math', data=data, palette =color, hue='Athlete')
#distribution by state
sns.swarmplot(x='Gender', y='Math', data=data, palette =color, hue='State')
```

```python
sns.stripplot(x='Gender', y='Math', data=data, palette =color)
```

```python
sns.violinplot(x='Gender', y='Math', data=data, palette =color)
```
