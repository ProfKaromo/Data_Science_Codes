```python
from scipy.stats import f_oneway
import pandas
data =  pandas.read_csv("https://modcom.co.ke/data/datasets/exams.csv")
data.head(4)
```

```python
# H0 ( ): null hypothesis μ1 = μ2 = μ3 = ... = μk ( all the population means are equal)
# H1 ( ): alternative hypothesis at least one population mean is different from the rest
# Confidence 95% - Risk 5%.   Reject H0 when a Pvalue < 0.05
stats, pvalue = f_oneway(data['group1'], data['group2'], data['group3'])
print('P value ', pvalue)
if pvalue < 0.05:
   print('Reject the Null hypothesis')
   print('We can conclude that at least one population mean is different from the rest ')
else:
   print('Fail Reject the Null hypothesis')
   print(' all the population means are equal')
```
